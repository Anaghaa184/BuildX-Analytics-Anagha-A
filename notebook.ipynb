```python
import pandas as pd
df = pd.read_csv("startup_funding.csv")
df.head()
#displaying first 5 rows
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Rename columns for easier access
df.rename(columns={
    'Date dd/mm/yyyy': 'date',
    'Startup Name': 'startup_name',
    'Industry Vertical': 'industry',
    'SubVertical': 'sub_vertical',
    'City  Location': 'city',
    'Investors Name': 'investors',
    'InvestmentnType': 'investment_type',
    'Amount in USD': 'amount_usd'
}, inplace=True)

df.columns  # confirm new names
```

```python
# Fill missing text columns with 'Unknown'
df['industry'].fillna('Unknown', inplace=True)
df['sub_vertical'].fillna('Unknown', inplace=True)
df['city'].fillna('Unknown', inplace=True)
df['investors'].fillna('Unknown', inplace=True)
df['investment_type'].fillna('Unknown', inplace=True)
# Remove commas from amount column so it can be converted to numeric
df['amount_usd'] = df['amount_usd'].astype(str).str.replace(',', '', regex=False)

# Replace text like 'undisclosed' with NaN since it's not a number
df['amount_usd'] = df['amount_usd'].str.lower().replace('undisclosed', pd.NA)

# Now convert to numeric — non-convertible values become NaN
df['amount_usd'] = pd.to_numeric(df['amount_usd'], errors='coerce')


# Fill missing amount with 0
df['amount_usd'].fillna(0, inplace=True)

# confirm no missing values remain
df.isnull().sum()
```

```python
# Remove exact duplicate rows

df.drop_duplicates(inplace=True)
print("Rows after removing duplicates:", len(df))
```

```python
# strip() removes spaces
# title() used to capitalize first letter of each word
# lower() for investment type to keep it uniform

df['startup_name'] = df['startup_name'].str.strip().str.title()
df['industry'] = df['industry'].str.strip().str.title()
df['sub_vertical'] = df['sub_vertical'].str.strip().str.title()
df['city'] = df['city'].str.strip().str.title()
df['investors'] = df['investors'].str.strip()
df['investment_type'] = df['investment_type'].str.strip().str.lower()

df.head()
```

```python
# Convert date column to datetime format
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')

# Convert amount_usd column to numeric
df['amount_usd'] = pd.to_numeric(df['amount_usd'], errors='coerce').fillna(0)

df.dtypes  # confirm types
```

```python
# Extract year from date — useful for year-wise trend analysis
df['funding_year'] = df['date'].dt.year
df[['date', 'funding_year']].head()
```

```python
# Save the final cleaned dataframe to a new CSV file
df.to_csv('cleaned_dataset.csv', index=False)
print("Cleaned dataset saved successfully!")
print("Final shape:", df.shape)
```
